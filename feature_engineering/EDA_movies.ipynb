```python
# budget and box office comparison of top 10 movies
movies = df['rank'].head(10).astype(str).tolist()

df['budget'] = df['budget'].str.extract(r'(\d+)')
budget = df["budget"].head(10).astype(float).tolist()

df['box_office'] = df['box_office'].str.extract(r'(\d+)')
box_office = df["box_office"].head(10).astype(float).tolist()


fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(movies, budget, label='Budget', color='red', edgecolor='black')
ax.bar(movies, box_office, label='Box Office', color='blue', edgecolor='black', bottom=budget)

plt.xlabel('Movies',fontsize=15,fontweight= "bold")
plt.ylabel('Amount',fontsize=15,fontweight= "bold")
plt.title('Budget vs. Box Office Performance of Top 10 Rated Movies',fontsize=18, pad=20,fontweight= "bold")
plt.xticks(rotation=45)
plt.legend(title='Category', loc='upper right')
plt.show()
```

```python
# analyzing top 10 most and least popular film genre

# extracting each unique genre
real_gen = ""
for i in df['genre']:
    real_gen = real_gen+","+i
all_gen = real_gen.split(',')[1:]
unique_genres = list(set(all_gen))

# extracting count of each unique genere
data = []
label = []
for i in unique_genres:
    c = 0
    for j in df['genre']:
        if i in j:
            c+=1
    data.append(c)
    label.append(i)
combined = list(zip(data,label))

# getting and plotting most popular
sorted_combined = sorted(combined, key=lambda x: x[1])
data = [x[0] for i,x in enumerate(sorted_combined) if i < 10]
label = [x[1] for i,x in enumerate(sorted_combined) if i < 10]

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(15, 10))
ax1.pie(data,labels=label,colors=sns.color_palette('Paired'),autopct='%.2f%%',rotatelabels=True,pctdistance=.9,textprops={'fontsize': 12})
ax1.set_title('Top 10 Most Popular',pad=50,fontsize=15,fontweight='bold')

# getting and plotting least popular
data = [x[0] for i,x in enumerate(reversed(sorted_combined)) if i < 10]
label = [x[1] for i,x in enumerate(reversed(sorted_combined)) if i < 10]

ax2.pie(data,labels=label,colors=sns.color_palette('Paired'),autopct='%.2f%%',rotatelabels=True,pctdistance=.9,textprops={'fontsize': 12})
ax2.set_title('Top 10 Least Popular',pad=50,fontsize=15,fontweight='bold')

plt.suptitle('Top 10 Most and Least Popular Film Genre ',fontweight='bold',fontsize=18)
plt.tight_layout()
plt.show()


```

```python
# getting movies percentage by year
_21to40 = 0
_41to60 = 0
_61to80 = 0
_81to00 = 0
_01to22 = 0

df['year'].sort_values()
for i in df['year']:
    if i>1920 and i<=1940:
        _21to40+=1
    elif i>1940 and i<=1960:
        _41to60+=1
    elif i>1960 and i<=1980:
        _61to80+=1
    elif i>1980 and i<=2000:
        _81to00+=1
    else: 
        _01to22+=1

data = [_21to40, _41to60, _61to80, _81to00, _01to22]
label = ['1921 - 1940', '1941- 1960', '1961 - 1980', '1981 - 2000', '2001 - 2022']
plt.figure(figsize=(12,6))
plt.pie(data,labels=label,colors=sns.color_palette('Paired'),autopct='%.2f%%',textprops={'fontsize': 12})
plt.title("Percentage of Movies based on Year",fontweight='bold',fontsize=18)
plt.show()
```
